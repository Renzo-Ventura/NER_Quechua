# 2 PREPROCESANDO EL DATASET

```python
import pandas as pd
data = pd.read_csv("/content/drive/MyDrive/proyecto/ner/support/ner_dataset.csv")
```

```python
data
```

```python
data = data.fillna(method = "ffill")
```

```python
data.head(20)
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
data["Sentence #"] = LabelEncoder().fit_transform(data["Sentence #"] )
```

```python
data
```

```python
data.rename(columns={"Sentence #":"sentence_id","Word":"words","Tag":"labels"}, inplace =True)
```

```python
data
```

```python
data["labels"] = data["labels"].str.upper()
data
```

```python
X= data[["sentence_id","words"]]
Y =data["labels"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size =0.2)
```

```python
train_data = pd.DataFrame({"sentence_id":x_train["sentence_id"],"words":x_train["words"],"labels":y_train})
test_data = pd.DataFrame({"sentence_id":x_test["sentence_id"],"words":x_test["words"],"labels":y_test})
```

```python
train_data
```

```python
test_data
```

# 3 ETIQUETADO

```python
label = data["labels"].unique().tolist()
label
```

# 4 FINE TUNING MODEL

```python
!pip install simpletransformers
```

```python
from simpletransformers.ner import NERModel,NERArgs
```

```python
args = NERArgs()
args.num_train_epochs = 10
args.learning_rate = 1e-4
args.overwrite_output_dir =True
args.train_batch_size = 32
args.eval_batch_size = 32
```

NER [model](https://simpletransformers.ai/docs/ner-model/)

```python
model = NERModel('roberta', "ctoshio/QuechuaBerta" ,labels=label, args =args,use_cuda=True)
```

```python
model.train_model(train_data,eval_data = test_data,acc=accuracy_score)
```

```python
result, model_outputs, preds_list = model.eval_model(test_data)
```

```python
result
```
